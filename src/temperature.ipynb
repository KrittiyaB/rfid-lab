```python
import pandas as pd

df_no_temp = pd.read_csv('../data/RFID_DATA_SET_NOTEMP_CLEANED.csv')
df_temp = pd.read_csv('../data/RFID_DATA_SET_TEMP_CLEANED.csv')
```

```python
temp21 = []
temp25 = []
temp29 = []
temp33 = []
temp37 = []
r_temp21 = []
r_temp25 = []
r_temp29 = []
r_temp33 = []
r_temp37 = []

for t, r, i in zip(df_temp['Temperature'],df_temp['RSSI'],df_temp['Detect']):
    if t == 21:
        temp21.append(r)
        r_temp21.append(i)
    elif t == 25:
        temp25.append(r)
        r_temp25.append(i)
    elif t == 29:
        temp29.append(r)
        r_temp29.append(i)
    elif t == 33:
        temp33.append(r)
        r_temp33.append(i)
    elif t == 37:
        temp37.append(r)
        r_temp37.append(i)
```

```python
# Relation RSSI with Temperature
from statistics import mean

all_temp = [mean(temp21), mean(temp25), mean(temp29), mean(temp33), mean(temp37)]
print(mean(temp21), mean(temp25), mean(temp29), mean(temp33), mean(temp37))
all_temp
```

```python
# Relation Read Rate (%) with Temperature
allReadTemp = [mean(r_temp21)*100, mean(r_temp25)*100, mean(r_temp29)*100, mean(r_temp33)*100, mean(r_temp37)*100]
print(mean(r_temp21)*100, mean(r_temp25)*100, mean(r_temp29)*100, mean(r_temp33)*100, mean(r_temp37)*100)
allReadTemp
```
