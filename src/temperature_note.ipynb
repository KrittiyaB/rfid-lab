```python
import pandas as pd

df = pd.read_csv('../data/EXP_TEMPERATURE.csv')
```

```python
import matplotlib.pyplot as plt

X1 = df['Highest']
X2 = df['Lowest']
Y = df['Date']
plt.figure(figsize=(10,5))
# plt.plot(df['Date'], df['Highest'], label='Highest', color='steelblue', linewidth = 1, marker='o', markerfacecolor='steelblue', markersize=8.5)
# plt.plot(df['Date'], df['Lowest'], label='Lowest', color='skyblue', linewidth = 1, marker='o', markerfacecolor='skyblue', markersize=8.5)

plt.plot(range(len(Y)), X1, label='Highest', color='orangered', linewidth = 1, marker='o', markerfacecolor='orangered', markersize=8.5) # Plotting data
plt.plot(range(len(Y)), X2, label='Lowest', color='steelblue', linewidth = 1, marker='o', markerfacecolor='steelblue', markersize=8.5)

plt.xticks(range(len(Y)), Y) # Redefining x-axis labels

for i, v in enumerate(X1):
    plt.text(i, v+1, "%.1f ᵒC" %v, ha="center")
for i, v in enumerate(X2):
    plt.text(i, v+1, "%.1f ᵒC" %v, ha="center")

plt.ylim(0, 50)
plt.ylabel('Temperature(ᵒC)')
plt.savefig('../img/stamp_temperature.png', dpi=150)

plt.show()

# plt.title('RSSI X Temperature')
```
